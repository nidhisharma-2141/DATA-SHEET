# **Q5- REPRESENT THE NO. OF RECORDS ACCORDING TO THE STATES USING PIE CHART**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
df = pd.read_csv('/content/weather - Sheet1 (1).csv')

# Count the number of rows per state
state_counts = df['state'].value_counts()

# Create a pie chart
plt.figure(figsize=(10, 10))
plt.pie(state_counts, labels=state_counts.index, autopct='%1.0f%%', startangle=9240, colors=plt.get_cmap('tab20').colors)

# Add title
plt.title('Number of records per State')

# Show the plot
plt.show()



```

# **Q9-VISUALISE THE MAXIMUM AIR INDEX , STATE WISE USING HORIZONTAL BAR CHART IN PYTHON**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
df = pd.read_csv('/content/weather - Sheet1 (1).csv')

# Extract the state, maximum air index
categories = df['state']
max_air_index=df['air index']


# Create the bar chart
plt.figure(figsize=(14, 7))

# Plot Max Air Index
plt.barh(categories, max_air_index, color='skyblue', label='Max Air Index', alpha=0.7)

# Add titles and labels
plt.xlabel('Category')
plt.ylabel('Values')
plt.title('Max Air Index by Category')
plt.legend()

# Rotate x labels if needed for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

# **Q13-SHOW THE STATES WITH HUMIDITY >80 AND PRESSURE >101.2 USING SUBPLOT**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
df = pd.read_csv('/content/weather - Sheet1 (1).csv')

# Filter the DataFrame to include only rows where Humidity > 80 and Pressure > 100.5
filtered_df = df[(df['Rel Hum_%'] > 80) & (df['Press_kPa'] > 101.2)]

# Extract the states, humidity, and pressure from the filtered DataFrame
states = filtered_df['state']
humidity = filtered_df['Rel Hum_%']
pressure = filtered_df['Press_kPa']

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))

# Plot Humidity in the first subplot
ax1.bar(states, humidity, color='darkblue')
ax1.set_xlabel('State')
ax1.set_ylabel('Humidity (%)')
ax1.set_title('States with Humidity > 80%')
ax1.set_xticklabels(states, rotation=45, ha='right')

# Plot Pressure in the second subplot
ax2.plot(states, pressure, marker='o', color='orange', linestyle='-', linewidth=2, markersize=8)
ax2.set_xlabel('State')
ax2.set_ylabel('Pressure (hPa)')
ax2.set_title('Pressure > 100.5 hPa')
ax2.set_xticklabels(states, rotation=45, ha='right')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

# **Q17-Build plot bar for date/time against the wind speed using matplotlib.**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
df = pd.read_csv('/weather - Sheet1.csv')

# Extract the weather conditions and wind speeds

wind_speeds = df['Wind Speed_km/h']
visibility = df['Visibility_km']

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(visibility, wind_speeds, color='skyblue')

# Add titles and labels
plt.xlabel('visibility')
plt.ylabel('Wind Speed (km/h)')
plt.title('visibility by wind speed')

# Rotate x labels if needed for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()

```
